***
### <font color='blue'>4. Evaluation du module
***

Evaluation :

Lors du TP évalué (dernière séance), vous aurez à votre disposition un jeu de données textuelles annotées (relatif au domaine de la santé)

Vous devrez réaliser en 4h un cas d'usage métier dans un notebook Jupyter

Il est attandu que vous réalisiez les étapes suivantes : 

1. Présentez les données (Statistiques et Dataviz)
2. Pré-processing (pourquoi vous appliquez tel ou tel traitement sur les données ?) 
3. Testez plusieurs type de représentation des données
4. Testez plusieurs Algorithmes de ML pour répondre au cas d'usage métier
5. Présentez vos résultats et analysez les erreurs
6. Conclusions et perspectives d'amélioration
